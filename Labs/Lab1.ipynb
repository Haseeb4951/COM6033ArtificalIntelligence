```python
import numpy as np
from scipy import stats

speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

speed_array = np.array(speed)
average_speed = np.mean(speed_array)
median_speed = np.median(speed_array)
mode_result = stats.mode(speed_array)  
print("Mode result:", mode_result)

mode_speed = mode_result.mode  

print(f"Average speed: {average_speed}")
print(f"Median speed: {median_speed}")
print(f"Most common speed (Mode): {mode_speed}")
```

```python
import numpy as np
from statistics import variance
from scipy import stats

game_points = np.array([35, 56, 43, 59, 63, 79, 35, 41, 64, 43, 93, 60, 77, 24, 82])

dt_var = variance(game_points)
print("Sample variance:", round(dt_var, 2))
dt_std = np.std(game_points)  
print("Sample std.dev:", round(dt_std, 2))
dt_rng = np.max(game_points, axis=0) - np.min(game_points, axis=0)
print("Range:", dt_rng)
print("Quantiles:")
for val in [20, 80, 100]:
    dt_qntls = np.percentile(game_points, val)
    print(str(val) + "%", dt_qntls)

q75, q25 = np.percentile(game_points, [75, 25])
print("Inter quartile range:", q75 - q25)
```

```python
import numpy as np

ages = [5, 31, 43, 48, 50, 41, 7, 11, 15, 39, 80, 82, 32, 2, 8, 6, 25, 36, 27, 61, 31]

ages_array = np.array(ages)
percentile_75 = np.percentile(ages_array, 75)
percentile_90 = np.percentile(ages_array, 90)

print(f"The 75th percentile is: {percentile_75}")
print(f"The age that 90% of people are younger than is: {percentile_90}")
```
